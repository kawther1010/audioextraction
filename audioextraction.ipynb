```python
import librosa
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
audio_path = '1 Minute Timer Relaxing Music Lofi Fish Background(MP3_160K).mp3'
y, sr = librosa.load(audio_path)
```

```python
mfccs = librosa.feature.mfcc(y=y, sr=sr)
print("Shape of MFCCs:", mfccs.shape)
```

```python
# Scale MFCCs
scaler = StandardScaler()
mfccs_scaled = scaler.fit_transform(mfccs)
mfccs_scaled
```

```python
# MFCCs Aggregation
mfccs_aggregated = np.mean(mfccs_scaled, axis=1)
print("Aggregated MFCCs:", mfccs_aggregated)
```
